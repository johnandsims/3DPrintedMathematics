Harmonic wave of the form
$$s(z,t) = A sin(kz - \omega t + \phi)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#####################
#  Constants
#####################
A = 5   #Amplitude (m)
L = 25  #Lambda- wavelength (m)
T = 20  #Period (s)
nT = 3  # number of periods
nZ = 4  # number of wavelengths
Nz = 1000  # Total number of points in z
Nt = 500   # Total number of points in t
phi = 0    # phase constant

w = 2*np.pi / T
k = 2*np.pi / L
tMax = nT * T
zMax = nZ * L
t = np.linspace(0, tMax, Nt)
z = np.linspace(0, zMax, Nz)
s = np.zeros((Nz, Nt))

for c in range(Nt):
    s[:, c] = A * np.sin(k * z - w * t[c] + phi)

Pt = np.where(t > 20)[0][0] - 1
Pz = np.where(z > 60)[0][0] - 1

print(Pt, Pz)
st = s[Pz, :]

sz = s[:, Pt]
```

Surface plots and plots with constant time and position.

```python


# Plotting
plt.figure(figsize=(12, 5))


#Plot 1
# Plotting the wavefunction s(z,t) as a heatmap  constant z.
plt.subplot(1, 2, 1)
plt.imshow(s, extent=[0, tMax, 0, zMax], aspect='auto', cmap='viridis')
plt.colorbar(label='Amplitude')
plt.xlabel('Time, t')
plt.ylabel('Position, Z')
plt.title('Wave Function s(z,t)')

# Plotting the wavefunction at a constant z and a varying in  t
plt.subplot(1, 2, 2)
plt.plot(t, st, label=f's(z={z[Pz]:.2f}, t)')
plt.xlabel('Time, t')
plt.ylabel('Amplitude, s')
plt.title('Temporal variation of s(z,t) at constant distance, z')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 5))
#Plot 2
# Plotting the 3D surface
T, Z = np.meshgrid(t, z)
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(121, projection='3d')
surf = ax.plot_surface(T, Z, s, cmap='viridis', rstride=10, cstride=10, linewidth=0, antialiased=False)
ax.set_xlabel('Time, t')
ax.set_ylabel('position, z')
ax.set_zlabel('Amplitude, s')
ax.set_title('3D Surface of Wave Function s(z,t)')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# viewing angle
ax.view_init(elev=30, azim=10)


# Plotting the wavefunction at a constant t and a constant z
plt.subplot(122)
plt.plot(z,sz, label=f's(z, t = {t[Pt]:.2f})')
plt.xlabel('Position, z')
plt.ylabel('Amplitude, s')
plt.title('Spatial variation of s(z,t) at constant time, t')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()

```
